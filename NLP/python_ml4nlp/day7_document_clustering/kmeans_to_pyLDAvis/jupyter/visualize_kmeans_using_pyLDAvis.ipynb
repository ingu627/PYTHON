```python
import warnings
warnings.filterwarnings('ignore')

import sys
sys.path.append('../')
import kmeans_to_pyLDAvis
import soyclustering
```

```python
# Load data

import pickle

with open('../data/2016-10-20-news-bow.pkl', 'rb') as f:
    params = pickle.load(f)
    x = params['x']
    index2word = params['index2word']
```

```python
# kmeans

from soyclustering import SphericalKMeans
from sklearn.feature_extraction.text import TfidfTransformer

x_tfidf = TfidfTransformer().fit_transform(x)
kmeans = SphericalKMeans(n_clusters=200, verbose=1)
labels = kmeans.fit_predict(x_tfidf)
centers = kmeans.cluster_centers_
```

```python
# visualize
from kmeans_to_pyLDAvis import kmeans_to_prepared_data

prepared_data = kmeans_to_prepared_data(
    x,
    index2word,
    centers,
    labels,
    n_printed_words = 40,
    radius = 2
)
```

```python
import pyLDAvis

pyLDAvis.save_html(prepared_data, 'kmeans_pyLDAvis.html')
pyLDAvis.display(prepared_data)
```
